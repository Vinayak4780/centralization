```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings  # Updated import
from langchain.vectorstores import FAISS
import os
from langchain.memory import ConversationBufferMemory

# File path for DSEU admission text
TEXT_FILE_PATH = "merged_text.txt"
FAISS_INDEX_PATH = "faiss_index_dseu_admissions"

# Load text document
if not os.path.exists(TEXT_FILE_PATH):
    print(f"Error: File not found at {TEXT_FILE_PATH}")
    exit(1)

loader = TextLoader(TEXT_FILE_PATH, encoding="utf-8")
documents = loader.load()

# Split text into manageable chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=100)
texts = text_splitter.split_documents(documents)

# âœ… Use Free Hugging Face Embeddings (New Version)
embedding_model = HuggingFaceEmbeddings(model_name="BAAI/bge-base-en")  # Free model
vectorstore = FAISS.from_documents(texts, embedding_model)

# Save FAISS index
vectorstore.save_local(FAISS_INDEX_PATH)
print(f"âœ… FAISS index created successfully at '{FAISS_INDEX_PATH}'!")
```

```python
import os
from langchain.vectorstores import FAISS
from langchain.chains import ConversationalRetrievalChain
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.memory import ConversationBufferMemory
from langchain.llms import LlamaCpp  # âœ… Corrected import for consistency

# âœ… Load FAISS index (DSEU Admissions)
FAISS_INDEX_PATH = "faiss_index_dseu_admissions"

if not os.path.exists(FAISS_INDEX_PATH):
    print(f"âŒ Error: FAISS index not found at {FAISS_INDEX_PATH}")
    exit(1)

vectorstore = FAISS.load_local(
    FAISS_INDEX_PATH,
    embeddings=HuggingFaceEmbeddings(model_name="BAAI/bge-base-en"),
    allow_dangerous_deserialization=True  # âœ… Ensure safe deserialization
)

# âœ… Load Free Local LLM (Mistral, Llama-2, GPT4All)
LLM_PATH = "mistral-7b-instruct-v0.1.Q4_K_M.gguf"  # ðŸ”¹ Replace with actual model path

if not os.path.exists(LLM_PATH):
    print(f"âŒ Error: Model file not found at {LLM_PATH}")
    exit(1)

llm = LlamaCpp(model_path=LLM_PATH, n_ctx=2048,temperature=0.1,max_tokens=20)

# âœ… Initialize Chatbot with FAISS Retriever
retriever = vectorstore.as_retriever()

# âœ… Initialize Conversation Memory
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

# âœ… Initialize Chatbot with Conversational Retrieval (Context Aware)
qa = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=retriever,
    memory=memory  # ðŸ”¹ Adds buffer memory support
)

# âœ… CLI Chatbot Loop
print("\nðŸŽ“ Welcome to the Free DSEU Admissions Chatbot ðŸ¤–")
print("ðŸ”¹ Ask any question about DSEU admissions.")
print("ðŸ”¹ Type 'exit' to end the chat.\n")

while True:
    query = input("You: ")  # User input
    if query.lower() in ["exit", "quit", "bye"]:
        print("Chatbot: Goodbye! ðŸ‘‹")
        break
    
    response = qa.run(query)
    print(f"Chatbot: {response}\n")
```
